# 🔍 Step 1: Data Loading & Exploration

This notebook explores the **Arvyax Reflective Dataset** to understand:
- Data shape, types, and missing values
- Target distributions (`emotional_state`, `intensity`)
- Feature distributions (categorical + numerical)
- Journal text characteristics
- Cross-feature relationships

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_colwidth', 80)
```

## 1. Load Data

```python
# ─────────────────────────────────────────
# 1. LOAD DATA
# ─────────────────────────────────────────
train = pd.read_csv('../data/train.csv')
test  = pd.read_csv('../data/test.csv')

print('=' * 50)
print('STEP 1: DATA LOADING & EXPLORATION')
print('=' * 50)
```

## 2. Basic Shape & Columns

```python
# ─────────────────────────────────────────
# 2. BASIC SHAPE & COLUMNS
# ─────────────────────────────────────────
print('\nSHAPE')
print(f'  Train : {train.shape}')   # (1200, 13)
print(f'  Test  : {test.shape}')    # (120, 11)

print('\nTRAIN COLUMNS:')
for col in train.columns:
    print(f'  - {col}')

print('\nTEST COLUMNS:')
for col in test.columns:
    print(f'  - {col}')

# Key observation: test has NO emotional_state or intensity
# Those are what you must PREDICT
```

## 3. Data Types

```python
# ─────────────────────────────────────────
# 3. DATA TYPES
# ─────────────────────────────────────────
print('\nDATA TYPES:')
print(train.dtypes)
```

## 4. Missing Values

```python
# ─────────────────────────────────────────
# 4. MISSING VALUES
# ─────────────────────────────────────────
print('\nMISSING VALUES IN TRAIN:')
missing = train.isnull().sum()
missing_pct = (missing / len(train) * 100).round(2)
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})
print(missing_df[missing_df['Missing Count'] > 0])

# Expected output:
# sleep_hours         7    0.58%
# previous_day_mood  15    1.25%
# face_emotion_hint 123   10.25%   <- biggest gap!

print('\nMISSING VALUES IN TEST:')
missing_test = test.isnull().sum()
print(missing_test[missing_test > 0])
```

## 5. Target Distributions

```python
# ─────────────────────────────────────────
# 5. TARGET DISTRIBUTIONS
# ─────────────────────────────────────────
print('\nTARGET: emotional_state')
print(train['emotional_state'].value_counts())
print(f'  -> Unique classes: {train["emotional_state"].nunique()}')
# calm(216), restless(209), neutral(201),
# focused(193), mixed(191), overwhelmed(190)
# -> BALANCED -- no need for class weighting

print('\nTARGET: intensity')
print(train['intensity'].value_counts().sort_index())
print(f'  -> Range: {train["intensity"].min()} to {train["intensity"].max()}')
# 1(226), 2(228), 3(240), 4(277), 5(229) -- BALANCED
```

## 6. Categorical Feature Distributions

```python
# ─────────────────────────────────────────
# 6. CATEGORICAL FEATURE DISTRIBUTIONS
# ─────────────────────────────────────────
cat_cols = ['ambience_type', 'time_of_day', 'previous_day_mood',
            'face_emotion_hint', 'reflection_quality']

print('\nCATEGORICAL FEATURES:')
for col in cat_cols:
    print(f'\n  [{col}]')
    print(train[col].value_counts(dropna=False))
```

## 7. Numerical Feature Stats

```python
# ─────────────────────────────────────────
# 7. NUMERICAL FEATURE STATS
# ─────────────────────────────────────────
num_cols = ['duration_min', 'sleep_hours', 'energy_level', 'stress_level']

print('\nNUMERICAL FEATURES:')
print(train[num_cols].describe().round(2))
```

## 8. Text Analysis

```python
# ─────────────────────────────────────────
# 8. TEXT ANALYSIS
# ─────────────────────────────────────────
print('\nJOURNAL TEXT ANALYSIS:')
train['text_length']     = train['journal_text'].str.len()
train['text_word_count'] = train['journal_text'].str.split().str.len()

print(train[['text_length', 'text_word_count']].describe().round(2))

# Find very short texts (potential problem cases)
short_texts = train[train['text_word_count'] <= 5]
print(f'\n  Warning: Very short texts (<=5 words): {len(short_texts)}')
print(short_texts[['journal_text', 'emotional_state', 'intensity']].head(5))
```

## 9. Cross-Analysis: State vs Key Features

```python
# ─────────────────────────────────────────
# 9. CROSS-ANALYSIS: STATE vs KEY FEATURES
# ─────────────────────────────────────────
print('\nAVG STRESS by emotional_state:')
print(train.groupby('emotional_state')['stress_level'].mean().round(2).sort_values(ascending=False))

print('\nAVG ENERGY by emotional_state:')
print(train.groupby('emotional_state')['energy_level'].mean().round(2).sort_values(ascending=False))

print('\nAVG INTENSITY by emotional_state:')
print(train.groupby('emotional_state')['intensity'].mean().round(2).sort_values(ascending=False))
```

## 10. Spot Check — Sample Rows

```python
# ─────────────────────────────────────────
# 10. SPOT CHECK — SAMPLE ROWS
# ─────────────────────────────────────────
print('\nSAMPLE ROWS:')
pd.set_option('display.max_colwidth', 60)
print(train[['journal_text', 'emotional_state',
             'intensity', 'stress_level', 'energy_level']].sample(5, random_state=42))

print('\nStep 1 Complete.')
```
